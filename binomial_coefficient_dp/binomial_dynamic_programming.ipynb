Pascal's triangle:
$$
\binom{n}{k} = \binom{n-1}{k-1} + \binom{n-1}{k}
$$

```python
def binom_recursive(n,k):
    # In the worst case recursion is called n/2 times, and in each call the recursion is called twice, 
    # thus the time complexity is O(2^n).
    if n <= k or k == 0:
        return 1
    
    return binom_recursive(n-1, k-1) + binom_recursive(n-1,k)
```

```python
def binom_dp(n,k):
    # Memory O(k), compute O(n*k)
    A = [1 for _ in range(k+1)]
    for n_i in range(2,n+1):
        for k_i in range(k,0, -1):
            if(k_i > n_i - 1):
                continue
            A[k_i] = A[k_i - 1] + A[k_i]
    return A[k]
```

## Comparision

```python
import timeit

Ns = [2*i for i in range(14)] # k = n/2
recursive_times = []
dp_times =  []
for n in Ns:
    print(n)
    recursive_time = timeit.timeit(lambda: binom_recursive(n, int(n/2)), number=1)
    dp_time = timeit.timeit(lambda: binom_dp(n, int(n/2)), number=1)
    dp_times.append(dp_time)
    recursive_times.append(recursive_time)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(Ns,recursive_times)
plt.plot(Ns,dp_times)
plt.show()
```
